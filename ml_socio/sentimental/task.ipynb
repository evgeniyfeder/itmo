```python
import unicodedata

porter=PorterStemmer()
stop_words=stopwords.words('english')

def normalize_string(q):
    for key,reg in regex.items():
        q = regex[key].sub("", q)
    for word in emoticons.keys():
        q = q.replace(word, emoticons[word])
    q = q.lower()
    for word in contraction_mapping.keys():
        q = q.replace(word, contraction_mapping[word])
    q = re.sub(r"[\-\"`@#$%^&*(|)/~\[\]{\}:;+,._='!?]+", " ", q)
    
    q = unicodedata.normalize('NFKD', q).encode('ascii', errors='ignore').decode('utf8', errors='ignore')
    q = re.sub(r'\b([b-hB-Hj-zJ-Z] )', ' ', q)
    q = re.sub(r'( [b-hB-Hj-zJ-Z])\b', ' ', q)
        
    q = ' '.join([porter.stem(word) for word in word_tokenize(q) if not word in stop_words])
    return q
```

```python
from sklearn.feature_extraction import text
from numpy import append
from sklearn.pipeline import FeatureUnion,Pipeline

word_vectorizer = text.TfidfVectorizer(
          analyzer='word', ngram_range=(1, 3),
          min_df=2, use_idf=True, sublinear_tf=True)
char_vectorizer = text.TfidfVectorizer(
          analyzer='char', ngram_range=(3, 5),
          min_df=2, use_idf=True, sublinear_tf=True)
ngrams_vectorizer = Pipeline([('feats', FeatureUnion([('word_ngram', word_vectorizer),
                ('char_ngram', char_vectorizer),
                ])),])

test_df = pandas.read_csv("data/test.csv")
train_df = pandas.read_csv("data/train.csv")

from pandarallel import pandarallel
pandarallel.initialize()
train_df['cleaned_tweet']=train_df['TweetText'].parallel_apply(normalize_string)
test_df['cleaned_tweet']=test_df['TweetText'].parallel_apply(normalize_string)


vectorizer=ngrams_vectorizer.fit(train_df['cleaned_tweet'])


x_train=vectorizer.transform(train_df['cleaned_tweet'])
x_test=vectorizer.transform(test_df['cleaned_tweet'])
```
